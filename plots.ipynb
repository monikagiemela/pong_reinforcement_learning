```python
import json
import matplotlib.pyplot as plt

# Load new training logs
#with open("/mnt/data/latest_logs (1).json") as f:
with open("latest_logs.json") as f:
    logs_new = json.load(f)

episodes_new = logs_new["episodes"]
rewards_new = logs_new["rewards"]
avg_rewards_new = logs_new["avg_rewards"]
losses_new = logs_new["losses"]
epsilon_values_new = logs_new["epsilon_values"]

# Plot 1: Average rewards over episodes
plt.figure(figsize=(10,6))
plt.plot(episodes_new, avg_rewards_new, label="Average Reward (100-episode window)", color="blue")
plt.xlabel("Episodes")
plt.ylabel("Average Reward")
plt.title("Figure 1. Average reward progression across training episodes")
plt.legend()
plt.grid(True)
plt.show()

# Plot 2: Raw rewards over episodes
plt.figure(figsize=(10,6))
plt.plot(episodes_new, rewards_new, label="Episode Reward", color="green", alpha=0.6)
plt.xlabel("Episodes")
plt.ylabel("Reward")
plt.title("Figure 2. Distribution of episodic rewards across training")
plt.legend()
plt.grid(True)
plt.show()

# Plot 3: Training loss over episodes
plt.figure(figsize=(10,6))
plt.plot(episodes_new, losses_new, label="Loss", color="red", alpha=0.7)
plt.xlabel("Episodes")
plt.ylabel("Loss")
plt.title("Figure 3. Training loss measured across episodes")
plt.legend()
plt.grid(True)
plt.show()

# Plot 4: Epsilon decay
plt.figure(figsize=(10,6))
plt.plot(episodes_new, epsilon_values_new, label="Epsilon", color="purple")
plt.xlabel("Episodes")
plt.ylabel("Epsilon")
plt.title("Figure 4. Epsilon decay over training episodes")
plt.legend()
plt.grid(True)
plt.show()
```
